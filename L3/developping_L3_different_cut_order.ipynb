The weights are added as a key in the frames and are calculated as explained below.

For neutrino weights, one needs to multiply by the atmospheric neutrino flux. This flux depends on the energy and the theta angle of the neutrino. The weight is calculated as follows :

\begin{equation}
NuGen_{weight} = \frac{OneWeight \times \Phi_{atmo}}{NEvets \times n_{files}},
\end{equation}

the same algorithm is used to calculate the weight for Genie.

Corsika has to be weighted in the following way :

\begin{equation}
Corsica_{weight} = \frac{DiplopiaWeight \times Weight}{TimeScale \times n_{files}}
\end{equation}

The signal (wimpsim) weight is calculated as

\begin{equation}
Signal_{weight} = V_{gen} \times lep_{weight},
\end{equation}

where $V_{gen}$ is the generated volume and $lep_{weight}$ is the weight of the lepton caltculated in wimpsim.

```python
exp_livetime=1.0E-09*(exp[-1]['I3EventHeader'].start_time - exp[0]['I3EventHeader'].start_time)
weight = 1/exp_livetime
for expevent in exp :
    if not 'weight' in expevent:
        expevent['weight'] = dataclasses.I3Double(weight)
```

```python
import numpy as np
import matplotlib
from matplotlib import rc
import matplotlib.pyplot as plt
```

```python
# A function to make 1D histograms with cuts

def plotting_histo(data_type,key,cut,bins,log,color,wimp_weight):
    
    weightfactor = 1.
    if data_type == signal :
        weightfactor = 1.0E25*wimp_weight
    if data_type == corsika :
        weightfactor = 1
    if data_type == genie :
        weightfactor = 2
        
    KEY = key.split('[')[1].split(']')[0]
        
    hist([eval(key) for event in data_type if (eval(KEY) in event and eval(cut))], weights=[weightfactor*event['weight'].value for event in data_type if (eval(KEY) in event and eval(cut))], bins=bins,log=log,facecolor=color,histtype='step')
```

A nice feature to add is calculate 

\begin{equation}
\epsilon = \frac{NEvents_{BC}}{NEvents_{AC}}
\end{equation}

```python
# A function to calculate epsilon (after cuts)

def calculate_epsilon(data_type,cut):

    if data_type == signal :
        weightfactor = 1.0E25
    if data_type == corsika :
        weightfactor = 1
    if data_type == exp :
        weightfactor = 1
    if data_type == nugen :
        weightfactor = 1
    if data_type == genie :
        weightfactor = 2
    
    hist_content_bc, bin_edges_bc = np.histogram([event['SPEFit2'].dir.zenith for event in data_type if ('SPEFit2' in event and not isnan(event['SPEFit2'].dir.zenith))], weights=[weightfactor*event['weight'].value for event in data_type if ('SPEFit2' in event and not isnan(event['SPEFit2'].dir.zenith))], bins=arange(-1,3.5,0.1))
    hist_content_ac, bin_edges_ac = np.histogram([event['SPEFit2'].dir.zenith for event in data_type if ('SPEFit2' in event and not isnan(event['SPEFit2'].dir.zenith) and eval(cut))], weights=[weightfactor*event['weight'].value for event in data_type if ('SPEFit2' in event and not isnan(event['SPEFit2'].dir.zenith) and eval(cut))], bins=arange(-1,3.5,0.1))
    return float(sum(hist_content_ac))/float(sum(hist_content_bc)), sum(hist_content_bc), sum(hist_content_ac)
```

```python
def calculate_and_print_epsilon(data_type,cut):
    
    epsilon, rate_bc, rate_ac = calculate_epsilon(data_type,cut)

    if data_type == signal :
        type = 'signal'
    if data_type == corsika :
        type = 'corsika'
    if data_type == exp :
        type = 'exp'
    if data_type == nugen :
        type = 'nugen'
    if data_type == genie :
        type = 'genie'
        
    print '*',type
    print '   total rate before cuts : ',rate_bc,','#sum(hist_content_bc),','
    print '   total rate after cuts : ', rate_ac,','#sum(hist_content_ac),','
    print '   percentage passing of the cuts : ', 100*epsilon,'%'
```

```python
def plotting_histos(key='',cut='',bins=arange(0,1,0.1),min=0,max=10,title='',calc_efficiency=False,wimp_weight=1,simulation_only=False) :
    
    plt.figure(figsize=(16,8))
    plt.subplot(121)
    if not simulation_only : plotting_histo(exp,key,cut,bins,False,'red',wimp_weight)
    plotting_histo(corsika,key,cut,bins,False,'blue',wimp_weight)
    plotting_histo(nugen,key,cut,bins,False,'yellow',wimp_weight)
    plotting_histo(genie,key,cut,bins,False,'pink',wimp_weight)
    plotting_histo(signal,key,cut,bins,False,'green',wimp_weight)
    if not simulation_only : plt.legend(('exp','corsika','nugen','genie','signal','genie'),loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, fancybox=True, shadow=True)
    if simulation_only : plt.legend(('corsika','nugen','genie','signal'),loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, fancybox=True, shadow=True)
    plt.xlabel(title, fontsize=20)
    plt.ylabel('rates (Hz)', fontsize=15)
    plt.ylim([min,max])
    
    plt.subplot(122)
    if not simulation_only : plotting_histo(exp,key,cut,bins,True,'red',wimp_weight)
    plotting_histo(corsika,key,cut,bins,True,'blue',wimp_weight)
    plotting_histo(nugen,key,cut,bins,True,'yellow',wimp_weight)
    plotting_histo(genie,key,cut,bins,True,'pink',wimp_weight)
    plotting_histo(signal,key,cut,bins,True,'green',wimp_weight)
    if not simulation_only : plt.legend(('exp','corsika','nugen','genie','signal'),loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, fancybox=True, shadow=True)
    if simulation_only : plt.legend(('corsika','nugen','genie','signal'),loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, fancybox=True, shadow=True)
    plt.xlabel(title, fontsize=20)
    plt.ylabel('rates (Hz)', fontsize=15)
    plt.ylim([min,max])
    plt.show()
    
    if calc_efficiency :
        if not simulation_only : calculate_and_print_epsilon(exp,cut)
        calculate_and_print_epsilon(corsika,cut)
        calculate_and_print_epsilon(nugen,cut)
        calculate_and_print_epsilon(genie,cut)        
        calculate_and_print_epsilon(signal,cut)
```

Look at the efficiency $\eta$ as a function of the cut value

The function below calculates 

\begin{equation}
\eta = \frac{\epsilon_S}{\sqrt{\epsilon_B}} = \frac{\frac{NSignal_{AC}}{NSignal_{BC}}}{\sqrt{\frac{NBackground_{AC}}{NBackground_{BC}}}}
\end{equation}

```python
from matplotlib.patches import Ellipse

def draw_efficiencies(cut_key="", cut_relation="<", cut_start=0., cut_stop=1., cut_step=0.1, extra_cut="True"):
    
    efficiencies = []
    cut_values = []
    epsilons_signal = []
    rates_ac_background = []
    
    for cut_value in arange(cut_start, cut_stop, cut_step):
        cut_values.append(cut_value)
        epsilon_signal, rate_bc_signal, rate_ac_signal = calculate_epsilon(signal,cut_key+cut_relation+str(cut_value)+" and "+extra_cut)
        epsilon_background, rate_bc_background, rate_ac_background = calculate_epsilon(exp,cut_key+cut_relation+str(cut_value)+" and "+extra_cut)
        epsilons_signal.append(epsilon_signal)
        rates_ac_background.append(rate_ac_background)
        efficiency = epsilon_signal/math.sqrt(epsilon_background)
        efficiencies.append(efficiency)
    
    plt.figure(figsize=(16,8))
    
    plt.subplot(121)
    plt.plot(cut_values,efficiencies)
    plt.xlabel('cut value', fontsize=15)
    plt.ylabel(r'$\eta$', fontsize=20)
    
    plt.subplot(122)
    plt.scatter(epsilons_signal,rates_ac_background)
    labels = ['cut value = {0}'.format(cut_value) for cut_value in arange(cut_start, cut_stop, cut_step)]
    for label, x, y in zip(labels, epsilons_signal, rates_ac_background):
        plt.annotate(
            label, 
            xy = (x, y), xytext=(0, -45), size=12,
            textcoords = 'offset points', #ha = 'right', va = 'bottom',
            bbox=dict(boxstyle="round", fc=(1.0, 0.7, 0.7), ec="none"),
            arrowprops=dict(arrowstyle="wedge,tail_width=1.",
                                fc=(1.0, 0.7, 0.7), ec="none",
                                patchA=None,
                                patchB=Ellipse((2, -1), 0.5, 0.5),
                                relpos=(0.2, 0.5),
                                )
            )
    plt.xlabel(r'$\epsilon_S$', fontsize=20)
    plt.ylabel('background rate (Hz)', fontsize=15)
    
    plt.show()
```

```python
def make_ratio_plot(key='', cut='', bins=arange(0,1,0.1), which_simulation=corsika) :

    KEY = key.split('[')[1].split(']')[0]
    
    hist_content_exp, bin_edges_exp = np.histogram([eval(key) for event in exp if (eval(KEY) in event and eval(cut))], weights=[event['weight'].value for event in exp if (eval(KEY) in event and eval(cut))], bins=bins)
    hist_content_sim, bin_edges_sim = np.histogram([eval(key) for event in which_simulation if (eval(KEY) in event and eval(cut))], weights=[event['weight'].value for event in which_simulation if (eval(KEY) in event and eval(cut))], bins=bins)

    content_ratio_plot = np.array(hist_content_sim[:])/np.array(hist_content_exp[:])
    edges = np.array(bin_edges_exp)+(bin_edges_exp[1]-bin_edges_exp[0])/2
    
    hist(edges[0:-1],weights=content_ratio_plot, bins=len(content_ratio_plot))#, histtype='step')
```

```python
def plotting_2d_histo(data_type=exp, x_key='', y_key='', cut='True', bins=arange(100,100), simulation_only = False):

    X_KEY = x_key.split('[')[1].split(']')[0]
    Y_KEY = y_key.split('[')[1].split(']')[0]

    if simulation_only == False:
        weights = [event['weight'].value for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
        x = [eval(x_key) for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
        y = [eval(y_key) for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
    
    else :
        weights = [event['weight'].value for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
        x = [eval(x_key) for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
        y = [eval(y_key) for event in data_type if ((eval(X_KEY) in event) and (eval(Y_KEY) in event) and eval(cut))]
    
    h2d_weighted, xedges_weighted, yedges_weighted = np.histogram2d(x,y,bins=bins,weights=weights)
    plt.pcolormesh(xedges_weighted,yedges_weighted,h2d_weighted.T)
    plt.colorbar()
```

For the online filters, we keep VEF, LowUp and DC filtered events.

Now I make 3 cuts to reduce the background :

* a cut on position of the starting point, i.e. the finite reco $r = \sqrt{x^2 + y^2}$   and   $z$ . For signal, these starting posistions are mostly in the lower, central part from the detector, while for atmospheric muons, this is the opposite.
* a cut on the reconstructed zenith angle. Signal is vertically upgoing, while data is dominated by downgoing muons. The cut on zenth is made at 120 degrees, such that there is still data from outside the signal region that will serve as a control region for the monte carlo
* a cut on NDir/length. This should be big for signal, while it is small for muons, since low energy vertical tracks should have many direct hits (close to string and vertical DOM seperation is only 17m) and their track length will be small (low energy). The opposite is true for atmospheric muons.

Now that we made a veto for the starting point of the track, we can cut on the reconstructed zenith angle. It is important that this cut is not too hard, because we still want to keep an ofsource  control region to see if MC and dat match. for that reason, the cut is chosen at ~30 degrees below horizon, i.e.

\begin{equation}
\theta_{SPE} > 2.0 rad \sim 120^{\circ}
\end{equation}

since this number should always be bigger in DC events, because there you will have many direct hits on short length scales.
