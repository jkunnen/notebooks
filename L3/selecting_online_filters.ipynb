The weights are added as a key in the frames and are calculated as explained below.

For neutrino weights, one needs to multiply by the atmospheric neutrino flux. This flux depends on the energy and the theta angle of the neutrino. The weight is calculated as follows :

\begin{equation}
NuGen_{weight} = \frac{OneWeight \times \Phi_{atmo}}{NEvets \times n_{files}},
\end{equation}

the same algorithm is used to calculate the weight for Genie.

Corsika has to be weighted in the following way :

\begin{equation}
Corsica_{weight} = \frac{DiplopiaWeight \times Weight}{TimeScale \times n_{files}}
\end{equation}

The signal (wimpsim) weight is calculated as

\begin{equation}
Signal_{weight} = V_{gen} \times lep_{weight},
\end{equation}

where $V_{gen}$ is the generated volume and $lep_{weight}$ is the weight of the lepton caltculated in wimpsim.

```python
exp_livetime=1.0E-09*(exp[-1]['I3EventHeader'].start_time - exp[0]['I3EventHeader'].start_time)
weight = 1/exp_livetime
for expevent in exp :
    if not 'weight' in expevent:
        expevent['weight'] = dataclasses.I3Double(weight)
```

## Make the plotting functions

```python
import numpy as np
import matplotlib
from matplotlib import rc
import matplotlib.pyplot as plt
```

```python
# A function to calculate epsilon (after cuts)

def calculate_rate(data_type,cut):
    
    rate = sum([event['weight'].value for event in data_type if eval(cut)])
    return rate
```

```python
def get_rates(data_type):
    rate_before_filters = calculate_rate(data_type,'True')
    rate_after_VEF = calculate_rate(data_type,"event['FilterMask']['SDST_VEF_11'].condition_passed==True")
    rate_after_VEF_LowUp = calculate_rate(data_type,"(event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True)")
    rate_after_DC_VEF_LowUp = calculate_rate(data_type,"(event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True) ")
    rate_after_muon_DC_VEF = calculate_rate(data_type,"(event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True) ")
    rate_after_muon_DC_LowUp = calculate_rate(data_type,"(event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True)  or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True)")
    rate_after_muon_VEF_LowUp = calculate_rate(data_type,"(event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True) or (event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True)")
    rate_after_muon_DC_VEF_LowUp = calculate_rate(data_type,"(event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True) or (event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True) ")
    return rate_before_filters, [rate_after_VEF,rate_after_VEF_LowUp,rate_after_DC_VEF_LowUp,rate_after_muon_DC_VEF,rate_after_muon_DC_LowUp,rate_after_muon_VEF_LowUp,rate_after_muon_DC_VEF_LowUp]
```

```python
signal_rates={}
signal_efficiencies={}
for signal in [signal_m35_ch5,signal_m35_ch11,signal_m50_ch5,signal_m50_ch11,signal_m100_ch5,signal_m100_ch8,signal_m250_ch5,signal_m250_ch8,signal_m500_ch5,signal_m500_ch8,signal_m1000_ch5,signal_m1000_ch8,signal_m3000_ch5,signal_m3000_ch8,signal_m5000_ch5,signal_m5000_ch8]:
    signal_rates[str(signal)] = get_rates(signal)
    signal_efficiencies[str(signal)] = [rate/signal_rates[str(signal)][0]*100 for rate in signal_rates[str(signal)][1]]
```

```python
exp_rates   = get_rates(exp)
exp_efficiencies = [rate/exp_rates[0]*100 for rate in exp_rates[1]]
```
