The weights are added as a key in the frames and are calculated as explained below.

For neutrino weights, one needs to multiply by the atmospheric neutrino flux. This flux depends on the energy and the theta angle of the neutrino. The weight is calculated as follows :

\begin{equation}
NuGen_{weight} = \frac{OneWeight \times \Phi_{atmo}}{NEvets \times n_{files}},
\end{equation}

the same algorithm is used to calculate the weight for Genie.

Corsika has to be weighted in the following way :

\begin{equation}
Corsica_{weight} = \frac{DiplopiaWeight \times Weight}{TimeScale \times n_{files}}
\end{equation}

The signal (wimpsim) weight is calculated as

\begin{equation}
Signal_{weight} = V_{gen} \times lep_{weight},
\end{equation}

where $V_{gen}$ is the generated volume and $lep_{weight}$ is the weight of the lepton caltculated in wimpsim.

```python
def max_hitted_doms_per_string(pulses):
    strings = [key.string for key in pulses.keys()]
    n_strings = [strings.count(string) for string in strings]
    max_hitted_doms_per_string = max(n_strings)
    return max_hitted_doms_per_string
```

```python
def max_hit_doms_per_string(pulseseries):
    for data_type in [signal_m35_ch5,signal_m35_ch11,signal_m50_ch5,signal_m50_ch11,signal_m100_ch5,signal_m100_ch8,signal_m250_ch5,signal_m250_ch8,signal_m500_ch5,signal_m500_ch8,signal_m1000_ch5,signal_m1000_ch8,signal_m3000_ch5,signal_m3000_ch8,signal_m5000_ch5,signal_m5000_ch8]:
        for event in data_type :
            if pulseseries in event :
                pulses = event[pulseseries]
                if pulses.__class__ == dataclasses.I3RecoPulseSeriesMapMask:
                    pulses = pulses.apply(event)
                max_hit_doms_per_string = max_hitted_doms_per_string(pulses)
   
            if not 'max_hit_doms_per_string'+pulseseries in event :
                event['max_hit_doms_per_string'+pulseseries]=dataclasses.I3Double(max_hit_doms_per_string) 
```

```python
def calculate_rate(data_type,cut):

    rate = sum([event['weight'].value for event in data_type if eval(cut)])

    return rate
```

```python
def get_rates(data_type):
    cut_1='True'
    rate_before_filters = calculate_rate(data_type,cut_1)
    cut_2="((event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True))"
    rate_after_muon_DC_VEF = calculate_rate(data_type,cut_2)
    cut_3="((event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True))"
    rate_after_DC_VEF_LowUp = calculate_rate(data_type,cut_3)
    cut_4=cut_3+" and (('FiniteRecoFit' in event) and (event['FiniteRecoFit'].pos.z < 180) and (sqrt(event['FiniteRecoFit'].pos.x**2+event['FiniteRecoFit'].pos.y**2)<490))"
    rate_after_FiniteReco_cut = calculate_rate(data_type,cut_4)
    cut_5=cut_4+" and (('SPEFit2'in event) and (event['SPEFit2'].dir.zenith > 2.0))"
    rate_after_zenith_cut = calculate_rate(data_type,cut_5)
    cut_6=cut_5+" and (event['HitStatisticsValues'].z_travel > 10)"
    rate_after_ztravel_cut = calculate_rate(data_type,cut_6)
    cut_7=cut_6+" and (   (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True) or \
          (  (event['FilterMask']['DeepCoreFilter_11'].condition_passed==False) and ( (event['NDirB_over_LDirB'].value > 0.03) and (event['NDirB_over_LDirB'].value < 0.09) )  )   )"
    rate_after_NDiroverLDir_cut = calculate_rate(data_type,cut_7)
    return rate_before_filters, [rate_after_muon_DC_VEF,rate_after_DC_VEF_LowUp,rate_after_FiniteReco_cut,rate_after_zenith_cut,rate_after_ztravel_cut,rate_after_NDiroverLDir_cut]
```

```python
exp_rates   = get_rates(exp)
exp_efficiencies = [rate/exp_rates[0]*100 for rate in exp_rates[1]]
```

```python
signal_rates={}
signal_efficiencies={}
for signal in [signal_m35_ch5,signal_m35_ch11,signal_m50_ch5,signal_m50_ch11,signal_m100_ch5,signal_m100_ch8,signal_m250_ch5,signal_m250_ch8,signal_m500_ch5,signal_m500_ch8,signal_m1000_ch5,signal_m1000_ch8,signal_m3000_ch5,signal_m3000_ch8,signal_m5000_ch5,signal_m5000_ch8]:
    signal_rates[str(signal)] = get_rates(signal)
    signal_efficiencies[str(signal)] = [rate/signal_rates[str(signal)][0]*100 for rate in signal_rates[str(signal)][1]]
```

```python
def get_rates(data_type):
    cut_1='True'
    rate_before_filters = calculate_rate(data_type,cut_1)
#    cut_2="((event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_MuonFilter_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True))"
#    rate_after_muon_DC_VEF = calculate_rate(data_type,cut_2)
#    cut_3="((event['FilterMask']['SDST_VEF_11'].condition_passed==True) or (event['FilterMask']['SDST_LowUp_11'].condition_passed==True) or (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True))"
#    rate_after_DC_VEF_LowUp = calculate_rate(data_type,cut_3)
    cut_4="(('FiniteRecoFit' in event) and (event['FiniteRecoFit'].pos.z < 180) and (sqrt(event['FiniteRecoFit'].pos.x**2+event['FiniteRecoFit'].pos.y**2)<490))"
    rate_after_FiniteReco_cut = calculate_rate(data_type,cut_4)
    cut_5=cut_4+" and (('SPEFit2'in event) and (event['SPEFit2'].dir.zenith > 2.0))"
    rate_after_zenith_cut = calculate_rate(data_type,cut_5)
    cut_6=cut_5+" and (event['HitStatisticsValues'].z_travel > 10)"
    rate_after_ztravel_cut = calculate_rate(data_type,cut_6)
    cut_7=cut_6+" and (   (event['FilterMask']['DeepCoreFilter_11'].condition_passed==True) or \
          (  (event['FilterMask']['DeepCoreFilter_11'].condition_passed==False) and ( (event['NDirB_over_LDirB'].value > 0.03) and (event['NDirB_over_LDirB'].value < 0.09) )  )   )"
    rate_after_NDiroverLDir_cut = calculate_rate(data_type,cut_7)
    return rate_before_filters, [rate_after_FiniteReco_cut,rate_after_zenith_cut,rate_after_ztravel_cut,rate_after_NDiroverLDir_cut]
```
