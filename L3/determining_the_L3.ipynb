## Loading the files and preparing everything

```python
from I3Tray import *
from icecube import dataclasses, icetray, filter_tools, phys_services, dataio, common_variables

import os, sys, glob

import numpy as np
import matplotlib
from matplotlib import rc
import matplotlib.pyplot as plt
import tables
from tables import *
```

```python
files = glob.glob('/ice3/jkunnen/cut_information_second_try*')

information = {}

for filename in files:
    info_file = tables.openFile(filename)
    
    information['signal_efficiency'+str(filename)]          = info_file.root.cut_information.cut_information.cols.signal_efficiency[:]
    information['backrgound_rate_after_cuts'+str(filename)] = info_file.root.cut_information.cut_information.cols.backrgound_rate_after_cuts[:]
    information['eta'+str(filename)]                        = info_file.root.cut_information.cut_information.cols.eta[:]   
    information['z_travel_cut'+str(filename)]               = info_file.root.cut_information.cut_information.cols.z_travel_cut[:]
    information['r_finite_reco_cut'+str(filename)]          = info_file.root.cut_information.cut_information.cols.r_finite_reco_cut[:]
    information['z_finite_reco_cut'+str(filename)]          = info_file.root.cut_information.cut_information.cols.z_finite_reco_cut[:]
    information['NDirB_over_LDirB_lower_cut'+str(filename)] = info_file.root.cut_information.cut_information.cols.NDirB_over_LDirB_lower_cut[:]
    information['NDirB_over_LDirB_upper_cut'+str(filename)] = info_file.root.cut_information.cut_information.cols.NDirB_over_LDirB_upper_cut[:]

    high_signal_efficiency_cut = information['signal_efficiency'+str(filename)] > 0.40
    low_background_rate_cut = information['backrgound_rate_after_cuts'+str(filename)] < 1.
    very_low_background_rate_cut = information['backrgound_rate_after_cuts'+str(filename)] < 0.82
    high_signal_efficiency_low_background_rate_cut = high_signal_efficiency_cut*low_background_rate_cut
    high_signal_efficiency_very_low_background_rate_cut = high_signal_efficiency_cut*very_low_background_rate_cut
    
    information['hs_le_signal_efficiency'+str(filename)]          = information['signal_efficiency'+str(filename)][high_signal_efficiency_low_background_rate_cut]
    information['hs_le_backrgound_rate_after_cuts'+str(filename)] = information['backrgound_rate_after_cuts'+str(filename)][high_signal_efficiency_low_background_rate_cut] 
    information['hs_le_eta'+str(filename)]                        = information['eta'+str(filename)][high_signal_efficiency_low_background_rate_cut]                        
    information['hs_le_z_travel_cut'+str(filename)]               = information['z_travel_cut'+str(filename)][high_signal_efficiency_low_background_rate_cut]               
    information['hs_le_r_finite_reco_cut'+str(filename)]          = information['r_finite_reco_cut'+str(filename)][high_signal_efficiency_low_background_rate_cut]          
    information['hs_le_z_finite_reco_cut'+str(filename)]          = information['z_finite_reco_cut'+str(filename)][high_signal_efficiency_low_background_rate_cut]          
    information['hs_le_NDirB_over_LDirB_lower_cut'+str(filename)] = information['NDirB_over_LDirB_lower_cut'+str(filename)][high_signal_efficiency_low_background_rate_cut] 
    information['hs_le_NDirB_over_LDirB_upper_cut'+str(filename)] = information['NDirB_over_LDirB_upper_cut'+str(filename)][high_signal_efficiency_low_background_rate_cut] 
    
    information['hs_very_le_signal_efficiency'+str(filename)]          = information['signal_efficiency'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]
    information['hs_very_le_backrgound_rate_after_cuts'+str(filename)] = information['backrgound_rate_after_cuts'+str(filename)][high_signal_efficiency_very_low_background_rate_cut] 
    information['hs_very_le_eta'+str(filename)]                        = information['eta'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]                        
    information['hs_very_le_z_travel_cut'+str(filename)]               = information['z_travel_cut'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]               
    information['hs_very_le_r_finite_reco_cut'+str(filename)]          = information['r_finite_reco_cut'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]          
    information['hs_very_le_z_finite_reco_cut'+str(filename)]          = information['z_finite_reco_cut'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]          
    information['hs_very_le_NDirB_over_LDirB_lower_cut'+str(filename)] = information['NDirB_over_LDirB_lower_cut'+str(filename)][high_signal_efficiency_very_low_background_rate_cut] 
    information['hs_very_le_NDirB_over_LDirB_upper_cut'+str(filename)] = information['NDirB_over_LDirB_upper_cut'+str(filename)][high_signal_efficiency_very_low_background_rate_cut]
    
    info_file.close()
    
information['signal_efficiency']=np.array([signal_efficiency for filename in files for signal_efficiency in information['signal_efficiency'+str(filename)][:]])
information['backrgound_rate_after_cuts']=[backrgound_rate_after_cuts for filename in files for backrgound_rate_after_cuts in information['backrgound_rate_after_cuts'+str(filename)]]
information['eta']=[eta for filename in files for eta in information['eta'+str(filename)]]
information['z_travel_cut']=[z_travel_cut for filename in files for z_travel_cut in information['z_travel_cut'+str(filename)]]
information['r_finite_reco_cut']=[r_finite_reco_cut for filename in files for r_finite_reco_cut in information['r_finite_reco_cut'+str(filename)]]     
information['z_finite_reco_cut']=[z_finite_reco_cut for filename in files for z_finite_reco_cut in information['z_finite_reco_cut'+str(filename)]]       
information['NDirB_over_LDirB_lower_cut']=[NDirB_over_LDirB_lower_cut for filename in files for NDirB_over_LDirB_lower_cut in information['NDirB_over_LDirB_lower_cut'+str(filename)]] 
information['NDirB_over_LDirB_upper_cut']=[NDirB_over_LDirB_upper_cut for filename in files for NDirB_over_LDirB_upper_cut in information['NDirB_over_LDirB_upper_cut'+str(filename)]]

information['hs_le_signal_efficiency']=np.array([signal_efficiency for filename in files for signal_efficiency in information['hs_le_signal_efficiency'+str(filename)][:]])
information['hs_le_backrgound_rate_after_cuts']=[backrgound_rate_after_cuts for filename in files for backrgound_rate_after_cuts in information['hs_le_backrgound_rate_after_cuts'+str(filename)]]
information['hs_le_eta']=[eta for filename in files for eta in information['hs_le_eta'+str(filename)]]
information['hs_le_z_travel_cut']=[z_travel_cut for filename in files for z_travel_cut in information['hs_le_z_travel_cut'+str(filename)]]
information['hs_le_r_finite_reco_cut']=[r_finite_reco_cut for filename in files for r_finite_reco_cut in information['hs_le_r_finite_reco_cut'+str(filename)]]     
information['hs_le_z_finite_reco_cut']=[z_finite_reco_cut for filename in files for z_finite_reco_cut in information['hs_le_z_finite_reco_cut'+str(filename)]]       
information['hs_le_NDirB_over_LDirB_lower_cut']=[NDirB_over_LDirB_lower_cut for filename in files for NDirB_over_LDirB_lower_cut in information['hs_le_NDirB_over_LDirB_lower_cut'+str(filename)]] 
information['hs_le_NDirB_over_LDirB_upper_cut']=[NDirB_over_LDirB_upper_cut for filename in files for NDirB_over_LDirB_upper_cut in information['hs_le_NDirB_over_LDirB_upper_cut'+str(filename)]]

information['hs_very_le_signal_efficiency']=np.array([signal_efficiency for filename in files for signal_efficiency in information['hs_very_le_signal_efficiency'+str(filename)][:]])
information['hs_very_le_backrgound_rate_after_cuts']=[backrgound_rate_after_cuts for filename in files for backrgound_rate_after_cuts in information['hs_very_le_backrgound_rate_after_cuts'+str(filename)]]
information['hs_very_le_eta']=[eta for filename in files for eta in information['hs_very_le_eta'+str(filename)]]
information['hs_very_le_z_travel_cut']=[z_travel_cut for filename in files for z_travel_cut in information['hs_very_le_z_travel_cut'+str(filename)]]
information['hs_very_le_r_finite_reco_cut']=[r_finite_reco_cut for filename in files for r_finite_reco_cut in information['hs_very_le_r_finite_reco_cut'+str(filename)]]     
information['hs_very_le_z_finite_reco_cut']=[z_finite_reco_cut for filename in files for z_finite_reco_cut in information['hs_very_le_z_finite_reco_cut'+str(filename)]]       
information['hs_very_le_NDirB_over_LDirB_lower_cut']=[NDirB_over_LDirB_lower_cut for filename in files for NDirB_over_LDirB_lower_cut in information['hs_very_le_NDirB_over_LDirB_lower_cut'+str(filename)]] 
information['hs_very_le_NDirB_over_LDirB_upper_cut']=[NDirB_over_LDirB_upper_cut for filename in files for NDirB_over_LDirB_upper_cut in information['hs_very_le_NDirB_over_LDirB_upper_cut'+str(filename)]]
```
