```python
livetime = (max(exp.root.I3EventHeader.cols.time_start_mjd_day[:] * 24 * 3600 + exp.root.I3EventHeader.cols.time_start_mjd_sec[:] + exp.root.I3EventHeader.cols.time_start_mjd_ns[:] * 10 **-9) - min(exp.root.I3EventHeader.cols.time_start_mjd_day[:] * 24 * 3600 + exp.root.I3EventHeader.cols.time_start_mjd_sec[:] + exp.root.I3EventHeader.cols.time_start_mjd_ns[:] * 10 **-9))
```

```python
Exp['weight']= np.ones(len(exp.root.SPEFit2.cols.zenith[:]))/livetime
Wimp['weight']=wimp.root.weight.cols.value[:]
Corsika['weight']=corsika.root.weight.cols.value[:]
Nugen['weight']=nugen.root.weight.cols.value[:]
```

```python
def histWithError(data,bins=10,range=(0,100),weights=None,histtype='step',label=None,color="k",linestyle='solid',timescale=None):
    hist_content, bin_edges = np.histogram(data,bins=bins,range=range)
    plt.hist(data,weights=weights,bins=bins,range=range,histtype=histtype,label=label,color=color,linestyle=linestyle,alpha=0)
    bin_center = np.array(bin_edges)+(bin_edges[1]-bin_edges[0])/2
    binerrhz = np.sqrt(hist_content)/timescale
    plt.errorbar(bin_center[0:-1],hist_content/timescale,yerr=binerrhz,ecolor=color,fmt=None,capsize=4,elinewidth=None)
```

```python
def plot_1d_histogram(exp=[[],[]], corsika=[[],[]], nugen=[[],[]], wimp=[[],[]], bins=arange(-5,185,5),xlabel=r'$\rm{zenith \ angle \ (degrees)}$',xlim=[0,180],ylim=[0.001,10],axvline=None,axhline=None,**kwargs) :
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(25,10))
    
    n, bins, patches = ax1.hist(exp[0], \
                                bins=bins, facecolor='red', alpha=0, \
                                weights=exp[1],**kwargs)
    ax1.plot(bins[1:], n, 'r', linewidth=2)
    
    n, bins, patches = ax1.hist(corsika[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=corsika[1],**kwargs)
    ax1.plot(bins[1:], n, 'b-.', linewidth=3)
    
    n, bins, patches = ax1.hist(nugen[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=nugen[1],**kwargs)
    ax1.plot(bins[1:], n, 'k:', linewidth=3)
    
    n, bins, patches = ax1.hist(wimp[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=wimp[1],**kwargs)
    ax1.plot(bins[1:], n, 'g--', linewidth=3)
    
    
    ax1.set_xlabel(xlabel,size=33)
    ax1.set_ylabel(r'$\rm{rate (Hz)}$',size=33)
    ax1.set_ylim(ylim)
    ax1.set_xlim(xlim)    
    
    n, bins, patches = ax2.hist(exp[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=exp[1],log=True,**kwargs)
    ax2.plot(bins[1:], n, 'r', linewidth=2,label=r'$\rm{experimtental \ data}$')
    
    n, bins, patches = ax2.hist(corsika[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=corsika[1],log=True,**kwargs)
    ax2.plot(bins[1:], n, 'b-.', linewidth=3,label=r'$\rm{atmospheric}  \ \mu \ (\rm{CORSIKA})$')
    
    n, bins, patches = ax2.hist(nugen[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=nugen[1],log=True,**kwargs)
    ax2.plot(bins[1:], n, 'k:', linewidth=3,label=r'$\rm{atmospheric}  \ \nu \ \rm{induced} \ \mu$')
    
    n, bins, patches = ax2.hist(wimp[0], \
                                bins=bins, facecolor='red', alpha=0,\
                                weights=wimp[1],log=True,**kwargs)
    ax2.plot(bins[1:], n, 'g--', linewidth=3,label=r'$\rm{signal} \ \nu \ \rm{induced} \ \mu$')
        
    ax2.set_xlabel(xlabel,size=33)
    ax2.set_ylabel(r'$\rm{rate (Hz)}$',size=33)
    ax2.set_ylim(ylim)
    ax2.set_xlim(xlim)
    
    if axvline :
        ax1.axvline(axvline,color='k',linewidth=2.5)
        ax2.axvline(axvline,color='k',linewidth=2.5)
    if axhline : 
        ax1.axvline(axhline,color='k',linewidth=2.5)
        ax2.axvline(axhline,color='k',linewidth=2.5)
        
    lgd = ax2.legend(loc='upper center', bbox_to_anchor=(-0.2, 1.27), ncol=2, fancybox=True, shadow=True)
```

```python
exp_data     = [ [zenith*180/pi for zenith in exp.root.SPEFit2.cols.zenith[:][(exp.root.SPEFit2.cols.zenith[:]<=pi)]] , 
                 Exp['weight'][(exp.root.SPEFit2.cols.zenith[:]<=pi)] ]

corsika_data = [ [zenith*180/pi for zenith in corsika.root.SPEFit2.cols.zenith[:][(corsika.root.SPEFit2.cols.zenith[:]<=pi)]] , 
                 Corsika['weight'][(corsika.root.SPEFit2.cols.zenith[:]<=pi)] ]

nugen_data   = [ [zenith*180/pi for zenith in nugen.root.SPEFit2.cols.zenith[:][(nugen.root.SPEFit2.cols.zenith[:]<=pi)]] , 
                 Nugen['weight'][(nugen.root.SPEFit2.cols.zenith[:]<=pi)] ]

wimp_data    = [ [zenith*180/pi for zenith in wimp.root.SPEFit2.cols.zenith[:][(wimp.root.SPEFit2.cols.zenith[:]<=pi)]] , 
                np.array([weight*10**24 for weight in Wimp['weight']])[(wimp.root.SPEFit2.cols.zenith[:]<=pi)] ]    

plot_1d_histogram(exp=exp_data, corsika=corsika_data, nugen=nugen_data, wimp=wimp_data, 
                  bins=arange(-5,185,5),xlabel=r'$\rm{zenith \ angle \ (degrees)}$',axvline=2/pi*180,
                  xlim=[0,180],ylim=[0.001,10])
```

```python
for datatype in exp, corsika, nugen, wimp :        
    
    FRr_cut = np.array([math.sqrt(frx**2+fry**2) for frx, fry in zip(datatype.root.FiniteRecoFit.cols.x[:],datatype.root.FiniteRecoFit.cols.y[:])]) < 490
    FRz_cut = datatype.root.FiniteRecoFit.cols.z[:] < 190
    zenith_cut = datatype.root.SPEFit2.cols.zenith[:] > 2.0
    ztravel_cut = datatype.root.HitStatisticsValues.cols.z_travel[:] > 10
    L3_cut = FRr_cut*FRz_cut*zenith_cut*ztravel_cut
    
    if datatype == exp : L3_cut_exp = L3_cut
    if datatype == corsika : L3_cut_corsika = L3_cut 
    if datatype == nugen : L3_cut_nugen = L3_cut 
    if datatype == wimp : L3_cut_wimp = L3_cut  
```

```python
def calculate_cut_efficiencis(cut_exp,cut_corsika,cut_nugen,cut_wimp) :
    
    print ' Exp. Data : \n   *rate before cut : ' + str(sum(Exp['weight'])) +' Hz  \n   *rate after cut :  '+ str(sum(Exp['weight'][cut_exp])) +' Hz \n   *cut efficiency : ' + str(sum(Exp['weight'][cut_exp])/sum(Exp['weight'])) + ' %'
    print ' Corsika : \n   *rate before cut : ' + str(sum(Corsika['weight'])) +' Hz  \n   *rate after cut :  '+ str(sum(Corsika['weight'][cut_corsika])) +' Hz \n   *cut efficiency : ' + str(sum(Corsika['weight'][cut_corsika])/sum(Corsika['weight'])) + ' %'
    print ' Nugen : \n   *rate before cut : ' + str(sum(Nugen['weight'])) +' Hz  \n   *rate after cut :  '+ str(sum(Nugen['weight'][cut_nugen])) +' Hz \n   *cut efficiency : ' + str(sum(Nugen['weight'][cut_nugen])/sum(Nugen['weight'])) + ' %'
    print ' Wimp : \n   *rate before cut : ' + str(sum(Wimp['weight'])) +' Hz  \n   *rate after cut :  '+ str(sum(Wimp['weight'][cut_wimp])) +' Hz \n   *cut efficiency : ' + str(sum(Wimp['weight'][cut_wimp])/sum(Wimp['weight'])) + ' %'
```

```python
calculate_cut_efficiencis(L3_cut_exp,L3_cut_corsika,L3_cut_nugen,L3_cut_wimp)
```
