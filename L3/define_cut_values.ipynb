## Loading the files and preparing everything

```python
from I3Tray import *
from icecube import dataclasses, icetray, filter_tools, phys_services, dataio, common_variables

import os, sys, glob

import numpy as np
import matplotlib
from matplotlib import rc
import matplotlib.pyplot as plt
```

```python
signal_files=glob.glob("/data/IC86/sim/2011/WimpSim/level2_Filtered/9010/FilteredLevel21_IC86.2011_wimp_earth_m50_ch11.009010.small.i3")

signal=[]
for signal_file in signal_files:
    signalfile = dataio.I3File(signal_file)
    while signalfile.more():
        signal.append(signalfile.pop_physics())
```

```python
exp_files=glob.glob("/ice3/jkunnen/data_laptop/i3/exp/FilteredLevel2_IC86.2011_data_Run00119612_small.i3")

exp=[]
for exp_file in exp_files:
    expfile = dataio.I3File(exp_file)
    while expfile.more():
        exp.append(expfile.pop_physics())
```

The weights are added as a key in the frames and are calculated as explained below.

For neutrino weights, one needs to multiply by the atmospheric neutrino flux. This flux depends on the energy and the theta angle of the neutrino. The weight is calculated as follows :

\begin{equation}
NuGen_{weight} = \frac{OneWeight \times \Phi_{atmo}}{NEvets \times n_{files}},
\end{equation}

the same algorithm is used to calculate the weight for Genie.

Corsika has to be weighted in the following way :

\begin{equation}
Corsica_{weight} = \frac{DiplopiaWeight \times Weight}{TimeScale \times n_{files}}
\end{equation}

The signal (wimpsim) weight is calculated as

\begin{equation}
Signal_{weight} = V_{gen} \times lep_{weight},
\end{equation}

where $V_{gen}$ is the generated volume and $lep_{weight}$ is the weight of the lepton caltculated in wimpsim.

```python
exp_livetime=1.0E-09*(exp[-1]['I3EventHeader'].start_time - exp[0]['I3EventHeader'].start_time)
weight = 1/exp_livetime
for expevent in exp :
    if not 'weight' in expevent:
        expevent['weight'] = dataclasses.I3Double(weight)
    else :
        expevent['weight'].value = weight
```

## Filling the arrays 

We will only use 5 variables for the L3 filter. For this, we make $Events$ that contain only the necessary information, i.e. 

* $z_{travel}$
* $r_{FiniteReco}$
* $z_{FiniteReco}$
* $\theta_{SPE}$
* $\frac{N_{Dir,B}}{L_{Dir,B}}$

we also keep the $weight $ of the event, so there are in total 5 keys in each event.

```python
class Event :
    def __init__(self):
        self.z_travel = None
        self.r_finite_reco = None
        self.z_finite_reco = None
        self.zenith = None
        self.NDirB_over_LDirB = None
        self.DC_filter = None
        self.weight = None
    def __str__(self):
        return 'Ztravel : %s, r_FiniteReco : %s, z_FiniteReco : %s, Zenith : %s, NDirB/LDirB : %s, DCFilter : %r, weight : %s'%(self.z_travel,self.r_finite_reco,self.z_finite_reco,self.zenith,self.NDirB_over_LDirB, self.DC_filter,self.weight)
```

We want to find the most optimal combination for the cut values. For this reason we put all this information in a class $CutValues$.

This class contains :

* lower cut on $z_{travel}$
* upper cut on $r_{FiniteReco}$
* upper cut on $z_{FiniteReco}$
* lower cut on $\frac{N_{Dir,B}}{L_{Dir,B}}$
* upper cut on $\frac{N_{Dir,B}}{L_{Dir,B}}$

```python
class CutValues :
    def __init__(self):
        self.z_travel_cut = None
        self.r_finite_reco_cut = None
        self.z_finite_reco_cut = None
        self.NDirB_over_LDirB_lower_cut = None
        self.NDirB_over_LDirB_upper_cut = None
    def __repr__(self):
        return '[ZTravel Cut : %s, r_FiniteReco Cut : %s, z_FiniteReco Cut : %s, NDirB/LDirB Lower Cut : %s, NDirB/LDirB Upper Cut : %s]'%(self.z_travel_cut,self.r_finite_reco_cut,self.z_finite_reco_cut,self.NDirB_over_LDirB_lower_cut,self.NDirB_over_LDirB_upper_cut)
```

Besides the cutvalues, there is ofcourse more interesting information. All the interesting information is stored in an $Information$ object.

This object contains :

* $CutValues$   (see above)
* the signal efficiency $\epsilon_S = \frac{\textrm{amount of signal events after cuts}}{\textrm{amount of signal events before cuts}}$
* the background efficiency $\epsilon_B = \frac{\textrm{amount of background events after cuts}}{\textrm{amount of background events before cuts}}$
* the background rate after the cuts
* $\eta = \frac{\epsilon_S}{\sqrt{\epsilon_B}}$

```python
class Information :
    def __init__(self):
        self.cut_values = CutValues()
        self.signal_efficiency = None
        self.background_efficiency = None
        self.backrgound_rate_after_cuts = None
        self.eta = None
    def __str__(self):
        return 'Cut Values : %s, \n   Signal Efficiency : %s,   Background Efficiency : %s,   Background Rate After Cuts : %s,   Eta : %s'%(self.cut_values, self.signal_efficiency, self.background_efficiency, self.backrgound_rate_after_cuts, self.eta)
```

Now we have all the tools. The next step is to fill the signal and background arrays.

```python
signal_events = []

for event in signal :

    signal_event = Event()
    
    if 'HitStatisticsValues' in event :
        signal_event.z_travel = event['HitStatisticsValues'].z_travel
    else :
        signal_event.z_travel = nan
        
    if 'FiniteRecoFit' in event :
        signal_event.r_finite_reco = sqrt(event['FiniteRecoFit'].pos.x**2+event['FiniteRecoFit'].pos.y**2)
    else :
        signal_event.r_finite_reco = nan
        
    if 'FiniteRecoFit' in event :
        signal_event.z_finite_reco = event['FiniteRecoFit'].pos.z
    else :
        signal_event.z_finite_reco = nan
    
    if 'SPEFit2' in event :
        signal_event.zenith = event['SPEFit2'].dir.zenith
    else : 
        signal_event.zenith = nan
    
    if 'SPEFit2DirectHitsB' in event and event['SPEFit2DirectHitsB'].dir_track_length>0 :
        signal_event.NDirB_over_LDirB = event['SPEFit2DirectHitsB'].n_dir_doms/event['SPEFit2DirectHitsB'].dir_track_length
    else :
        signal_event.NDirB_over_LDirB = nan
    
    signal_event.DC_filter = event['FilterMask']['DeepCoreFilter_11'].condition_passed
    
    signal_event.weight = event['weight'].value 
        
    signal_events.append(signal_event)
```

```python
exp_events = []

for event in exp :
    
    exp_event = Event()
    
    if 'HitStatisticsValues' in event :
        exp_event.z_travel = event['HitStatisticsValues'].z_travel
    else :
        exp_event.z_travel = nan
        
    if 'FiniteRecoFit' in event :
        exp_event.r_finite_reco = sqrt(event['FiniteRecoFit'].pos.x**2+event['FiniteRecoFit'].pos.y**2)
    else :
        exp_event.r_finite_reco = nan
        
    if 'FiniteRecoFit' in event :
        exp_event.z_finite_reco = event['FiniteRecoFit'].pos.z
    else :
        exp_event.z_finite_reco = nan
    
    if 'SPEFit2' in event :
        exp_event.zenith = event['SPEFit2'].dir.zenith
    else : 
        exp_event.zenith = nan
    
    if 'SPEFit2DirectHitsB' in event and event['SPEFit2DirectHitsB'].dir_track_length>0 :
        exp_event.NDirB_over_LDirB = event['SPEFit2DirectHitsB'].n_dir_doms/event['SPEFit2DirectHitsB'].dir_track_length
    else :
        exp_event.NDirB_over_LDirB = nan
    
    exp_event.DC_filter = event['FilterMask']['DeepCoreFilter_11'].condition_passed
    
    exp_event.weight = event['weight'].value
        
    exp_events.append(exp_event)
```

Now we calculate the weighted amount of events. 

For signal the weight is given by

\begin{equation}
Signal_{weight} = V_{gen} \times lep_{weight},
\end{equation}

where $V_{gen}$ is the generated volume and $lep_{weight}$ is the weight of the lepton caltculated in wimpsim.

And for experimental data, the weight is given by

\begin{equation}
Exp_{weight} = livetime^{-1},
\end{equation}

such that the total weight for the signal is just the total rate.

```python
signal_rate_before_cuts = 0
for event in signal_events :
    signal_rate_before_cuts+= event.weight
    
background_rate_before_cuts = 0
for event in exp_events :
    background_rate_before_cuts+= event.weight
```

```python
import tables
from tables import *
```

```python
class Info(IsDescription): 
    z_travel_cut               = Int32Col()
    r_finite_reco_cut          = Int32Col()
    z_finite_reco_cut          = Int32Col()
    NDirB_over_LDirB_lower_cut = Float32Col()
    NDirB_over_LDirB_upper_cut = Float32Col()
    signal_efficiency          = Float32Col()
    background_efficiency      = Float32Col()
    backrgound_rate_after_cuts = Float32Col()
    eta                        = Float32Col()
```

```python
h5_info_file = openFile("cut_information.h5", mode = "w", title = "Test file")
group = h5_info_file.createGroup("/", 'cut_information', 'Cut information')
table = h5_info_file.createTable(group, 'cut_information', Info, "Cut information")
info = table.row

for z_travel_cut in arange(-20,40,2):
    
    for r_finite_reco_cut in arange(200,500,30):
        
        for z_finite_reco_cut in arange(-300,200,40):
            
            for NDirB_over_LDirB_lower_cut in arange(0.0,0.04,0.01):
                
                for NDirB_over_LDirB_upper_cut in arange(0.07,0.15,0.01):
                    
                    signal_rate_after_cuts=0.
                    for event in signal_events :
                        if (event.zenith > 2.0) and (event.z_travel > z_travel_cut) and (event.r_finite_reco < r_finite_reco_cut) and (event.z_finite_reco < z_finite_reco_cut) and (event.DC_filter or ((event.DC_filter == False and (event.NDirB_over_LDirB >  NDirB_over_LDirB_lower_cut)) and (event.DC_filter == False and (event.NDirB_over_LDirB < NDirB_over_LDirB_upper_cut)))) :
                            signal_rate_after_cuts+=event.weight
                    
                    background_rate_after_cuts=0.
                    for event in exp_events :
                        if (event.zenith > 2.0) and (event.z_travel > z_travel_cut) and (event.r_finite_reco < r_finite_reco_cut) and (event.z_finite_reco < z_finite_reco_cut) and (event.DC_filter or ((event.DC_filter == False and (event.NDirB_over_LDirB >  NDirB_over_LDirB_lower_cut)) and (event.DC_filter == False and (event.NDirB_over_LDirB < NDirB_over_LDirB_upper_cut)))) :
                            background_rate_after_cuts+=event.weight
                    
                    info['z_travel_cut']  = z_travel_cut
                    info['r_finite_reco_cut']  = r_finite_reco_cut
                    info['z_finite_reco_cut']  = z_finite_reco_cut
                    info['NDirB_over_LDirB_lower_cut']  = NDirB_over_LDirB_lower_cut
                    info['NDirB_over_LDirB_upper_cut']  = NDirB_over_LDirB_upper_cut
                    info['signal_efficiency'] = signal_rate_after_cuts/signal_rate_before_cuts
                    info['background_efficiency'] = background_rate_after_cuts/background_rate_before_cuts
                    info['backrgound_rate_after_cuts'] = background_rate_after_cuts
                    info['eta'] = (signal_rate_after_cuts/signal_rate_before_cuts)/sqrt(background_rate_after_cuts/background_rate_before_cuts)
                    # Insert a new particle record
                    info.append()
                    
table.flush()
```
